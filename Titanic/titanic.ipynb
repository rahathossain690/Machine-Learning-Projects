```python
categorized = pd.get_dummies(dataset[['Pclass', 'Sex', 'Embarked']], drop_first=True)
dataset = dataset.drop(['Pclass', 'Sex', 'Embarked', 'Name', 'Ticket', 'Cabin'], axis=1)
dataset = pd.concat([dataset, categorized], axis=1)
dataset
```

```python
dataset = dataset.fillna(value={
    'Age': dataset['Age'].mean()
})
dataset
```

```python
from sklearn.model_selection import train_test_split
fit_data_X = dataset[['Age', 'SibSp', 'Parch', 'Fare', 'Pclass', 'Sex_male', 'Embarked_Q', 'Embarked_S']].to_numpy()
fit_data_Y = dataset[['Survived']].to_numpy().reshape(-1)
X_train, X_test, Y_train, Y_test = train_test_split(fit_data_X, fit_data_Y, test_size=0.2)
```

```python
models = []
names = []
accuracy = []

# Random forrest 
from sklearn.ensemble import RandomForestClassifier
models.append(('Random Forest', RandomForestClassifier(max_depth=10, random_state=True)))

# Logistic
from sklearn.linear_model import LogisticRegression
models.append(('Logistic', LogisticRegression(solver='liblinear', multi_class='ovr')))

# Linear Discriminant analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
models.append(('Linear Discriminant Analysis', LinearDiscriminantAnalysis()))
# KNN
from sklearn.neighbors import KNeighborsClassifier
models.append(('KNN', KNeighborsClassifier()))
# Decision tree
from sklearn.tree import DecisionTreeClassifier
models.append(('Decision Tree', DecisionTreeClassifier()))
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
models.append(('Gaussian Naive Bayes', GaussianNB()))
# SVM
from sklearn.svm import SVC
models.append(('SVM', SVC(gamma='auto')))
models
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

for name, model in models:
  kford = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
  score = cross_val_score(model, X_train, Y_train, cv=kford, scoring='accuracy')
  print('%s %f %f' % (name, score.mean(), score.std()))
```
