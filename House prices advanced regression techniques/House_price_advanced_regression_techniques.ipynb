```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
train_data = pd.read_csv("https://raw.githubusercontent.com/rahathossain690/Machine-Learning-Projects/master/House%20prices%20advanced%20regression%20techniques/train.csv")
print(train_data.shape)
train_data
```

```python
# reshaping all the categorial data
def reshaping_categorial_data(dataset):
  cat = ['MSSubClass', 'MSZoning', 'Street',
       'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
       'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
       'HouseStyle', 'OverallQual', 'OverallCond',
       'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
       'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
       'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType',
        'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC',
       'Fence', 'MiscFeature', 'SaleType',
       'SaleCondition']
  
  categorized = pd.get_dummies(dataset[cat])
  dataset = dataset.drop(cat, axis=1)
  dataset = pd.concat([dataset, categorized], axis=1)
  return dataset

train_data = reshaping_categorial_data(train_data)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # enables to show full data
    print(train_data.isna().sum())
```

```python
# remove nan data
def remove_nan(dataset):
  value_dict = {}
  for col in dataset.columns:
    value_dict[col] = dataset[col].mean()
  dataset = dataset.fillna(value=value_dict)
  return dataset

train_data = remove_nan(train_data)
print(train_data.shape)
```

```python
models = []

# Random forrest 
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
models.append(('Random Forest', RandomForestRegressor(max_depth=20, random_state=True)))

# Extra tree
models.append(('Extra tree', ExtraTreesRegressor(max_depth=20, random_state=True)))

# Logistic
from sklearn.linear_model import LinearRegression
models.append(('Logistic', LinearRegression()))

# Decision tree
from sklearn.tree import DecisionTreeRegressor
models.append(('Decision Tree', DecisionTreeRegressor()))

# SVM
from sklearn.svm import SVR
models.append(('SVR', SVR()))
models
```

```python
X = train_data.drop(['Id', 'SalePrice'], axis=1).to_numpy()
Y = train_data['SalePrice'].to_numpy().reshape(-1)
print(X.shape)

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2)

for name, model in models:
  model.fit(train_X, train_Y)
  pred = model.predict(test_X)
  err = mean_squared_error(test_Y, pred)
  print('%s %f' % (name, err))
```
