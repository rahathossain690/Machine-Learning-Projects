```python
# reshaping all the categorial data
def reshaping_categorial_data(dataset):
  cat = ['MSSubClass', 'MSZoning', 'Street',
       'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
       'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
       'HouseStyle', 'OverallQual', 'OverallCond',
       'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
       'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
       'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType',
        'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC',
       'Fence', 'MiscFeature', 'SaleType',
       'SaleCondition']
  
  categorized = pd.get_dummies(dataset[cat])
  dataset = dataset.drop(cat, axis=1)
  dataset = pd.concat([dataset, categorized], axis=1)
  return dataset

train_data = reshaping_categorial_data(train_data)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # enables to show full data
    print(train_data.isna().sum())
```

```python
# remove nan data
def remove_nan(dataset):
  value_dict = {}
  for col in dataset.columns:
    value_dict[col] = dataset[col].mean()
  dataset = dataset.fillna(value=value_dict)
  return dataset

train_data = remove_nan(train_data)
```

```python
models = []

# Random forrest 
from sklearn.ensemble import RandomForestClassifier
models.append(('Random Forest', RandomForestClassifier(max_depth=5, random_state=True)))

# Logistic
from sklearn.linear_model import LogisticRegression
models.append(('Logistic', LogisticRegression(solver='liblinear', multi_class='ovr')))

# Linear Discriminant analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
models.append(('Linear Discriminant Analysis', LinearDiscriminantAnalysis()))
# KNN
from sklearn.neighbors import KNeighborsClassifier
models.append(('KNN', KNeighborsClassifier()))
# Decision tree
from sklearn.tree import DecisionTreeClassifier
models.append(('Decision Tree', DecisionTreeClassifier()))
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
models.append(('Gaussian Naive Bayes', GaussianNB()))
# SVM
from sklearn.svm import SVC
models.append(('SVM', SVC(gamma='auto')))
models
```

```python
X = train_data.drop(['Id', 'SalePrice'], axis=1).to_numpy()
Y = train_data['SalePrice'].to_numpy().reshape(-1)

from sklearn.model_selection import cross_val_score, StratifiedKFold

for name, model in models:
  kford = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
  score = cross_val_score(model, X, Y, cv=kford, scoring='accuracy')
  print('%s %f %f' % (name, score.mean(), score.std()))
```
